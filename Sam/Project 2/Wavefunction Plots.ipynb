```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#These are the frequencies I test
w = [0.01, 0.05, 0.5, 1, 5, "Non-Interacting"]
#First energy eigenvalue from each case of omega.
#corresponding respectively with the array "w" above
E = [0.7145, 0.741, 2.1084, 2.9243, 4.5677, 0.4382]
i = 0
r = np.arange(0,15,0.1)
while i < len(E):
    #WF from APS M. Taut article
    wave = r*np.exp((E[i]/7)*-(r**2))
    plt.plot(r, wave, label = "w="+str(w[i]))
    i += 1
plt.title('Wavefunction with Varying Omega')
plt.xlabel('Radius')
plt.ylabel('Wavefunction')
plt.xlim([0,12])
plt.legend(loc='best')
plt.show()
```
